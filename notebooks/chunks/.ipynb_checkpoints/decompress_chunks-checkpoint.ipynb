# Library

```python
## Using the cmd and reading files
from os import popen, listdir
## Clear Output
from IPython.display import clear_output
```

```python
## Our worlds
worlds = ["overworld", "nether", "end"]
## Our prev_path for go to our destination
prev_path = "../../data/server/chunks/"
## A counter just because i like to see how slow my computer is
count = 0
```

```python
## Iterate for every worlds
for i in worlds:
    ## iterate for every files
    for chunks in listdir(prev_path + "32_" + i):
        ## Decompress the file
        popen(f"{prev_path}32_{i}/unix_decompression {prev_path}32_{i}/{chunks}")
        ## Print
        count += 1
        clear_output()
        print(count)
```
